```python
ds_a=ds[['first_name','balance','gender','age','amount','customer_id','month','day','hour']]
```

```python
ds_a=ds_a.replace({'gender':{'M':0,'F':1}})
```

```python
ds_a.info()
```

```python
ds_a_c=ds_a[['first_name','customer_id']]
```

```python
ds_a.drop(['first_name','customer_id'],inplace=True,axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
ds_ec=ds_a_c.apply(encoder.fit_transform)
```

```python
ds_a=pd.concat([ds_a,ds_ec],axis=1,sort=False)
ds_a.head()
```

```python
plt.figure(figsize=(12,10))
cor=ds_a.corr()
sns.heatmap(cor,annot=True,cmap=plt.cm.Reds)
plt.show()
```

```python
Y=ds_a['amount']
X=ds_a.drop(['amount'],axis=1)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
# print the intercept
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions = lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
Y = pd.get_dummies(Y)
```
